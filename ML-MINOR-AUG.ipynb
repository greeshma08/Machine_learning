Inorder to analyse the data ,we need to add a few more columns like total marks ,percentage and grades

```python
data['total marks']= data['math score'] + data['reading score'] + data['writing score']
```

```python
# since each subject are out of 100 so the total of all the subjects would be 300 
data['percentage'] = (data['total marks']/300)*100
```

Since in ethnicity we have grade A,grade B,grade C,grade D,grade E,we need to assign the grades according ,so:

 * 85-100 : Grade A
 * 70-84 : Grade B
 * 55-69 : Grade C
 * 35-54 : Grade D
 * 0-35 : Grade E
 

```python
# assigning the grade
def determine_grade(scores):
    if scores >= 85 and scores <= 100:
        return 'Grade A'
    elif scores >= 70 and scores < 85:
        return 'Grade B'
    elif scores >= 55 and scores < 70:
        return 'Grade C'
    elif scores >= 35 and scores < 55:
        return 'Grade D'
    elif scores >= 0 and scores < 35:
        return 'Grade E'
    
```

```python
data['grades']=data['percentage'].apply(determine_grade)
```

From the above result ,we find 3 new columns are added to the data set i.e total marks,percentage and grades

```python
data['grades'].value_counts().plot(kind='bar')
plt.show()
```
